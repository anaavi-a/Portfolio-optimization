```python
def CAGR(df,name):
    
    #USING df[name] and iloc function, identify the initial price of stock 
    start_value = df[name].iloc[0]
    
    #Like we did for Start Value, USING df[name] and iloc function, identify the final price of stock 
    end_value = df[name].iloc[-1]
    
    #Specify the Number of years we are taking into account 
    n = 3.67
    
    #WRITE THE CAGR FUNCTION USING START VALUE, END VALUE AND NUMBER OF PERIODS
    cagr = (end_value/start_value)**(1/n) -1 
    
    return cagr
```

```python
n = 1000
num_ports = n*50
num_of_assets = 50
all_weights = np.zeros((num_ports, num_of_assets))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for x in range(num_ports):
    if x%n == 0:
        print(x)
    # Weights
    weights = np.array(np.random.random(num_of_assets))
    weights = weights/np.sum(weights)
    # Save weights
    all_weights[x,:] = weights
    # Expected return
    ret_arr[x] = np.sum((returns_1.mean() * weights))
    #annualised_return => ret_arr[x] = np.sum((returns_1.mean() * weights*252))
    # Expected volatility
    vol_arr[x] = np.sqrt(np.dot(weights.T, np.dot(returns_1.cov(), weights)))
    #annualised_volatility => vol_arr[x] = np.sqrt(np.dot(weights.T, np.dot(returns_1.cov()*252, weights)))
    # Sharpe Ratio
    sharpe_arr[x] = ret_arr[x]/vol_arr[x]
```

```python
fy = np.linspace(0.00038,0.00062,100)
```

```python
def get_ret_vol_sr(weights):
    weights = np.array(weights)
    ret = np.sum(returns_1.mean() * weights) 
    vol = np.sqrt(np.dot(weights.T, np.dot(returns_1.cov(), weights)))
    sr = ret/vol
    return np.array([ret, vol, sr])

def min_vol(weights):
    return get_ret_vol_sr(weights)[1]

def check_sum(weights):
    #return 0 if sum of the weights is 1
    return np.sum(weights)-1
```

```python
a = 1/50
c = ((0,1),)
b = []
d = ()
for i in range(50):
    b.append(a)
    d = d+c
```

```python
from scipy.optimize import minimize
fx = []

for i in fy:
    print(i)
    offset = 0.00
    cons = ({'type':'eq', 'fun':check_sum},
            {'type':'eq', 'fun': lambda w: get_ret_vol_sr(w)[0] - i})
    
    result = minimize(min_vol, b, method='SLSQP', bounds=d, constraints=cons)
    fx.append(result['fun']+offset)   
```

```python
plt.figure(figsize=(15,8))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.plot(fx,fy, 'r--', linewidth=3)
plt.savefig('cover.png')
plt.show()
```
